```python
maximg = 19
i = 1
for state_name in df['State Name'].unique():
    temp = df[df['State Name'] == state_name]
    tempnew = temp.groupby('Quality Parameter').count()
    sns.barplot(x=tempnew.index, y= tempnew['State Name'], data = tempnew)
    plt.ylabel(state_name)
    plt.figure()
    if i == maximg:
        break
    i = i + 1
```

```python
import glob
import pandas as pd
import os
csvdir = 'india-csv'
allFiles = glob.glob(csvdir + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    try:
        df = pd.read_csv(file_,index_col=None, header=0)
        list_.append(df)
    except Exception as e:
        print("Error! in:", file_)
frame = pd.concat(list_)
```

```python
frame.to_csv("water_data.csv")
frame.fillna(0, inplace=True)
frame.head()
```
